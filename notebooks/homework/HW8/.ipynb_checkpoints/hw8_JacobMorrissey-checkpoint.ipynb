## Problem 2

```python
print('Problem 2:')
x = np.arange(-100,100,0.1)
y = np.arange(-100,100,0.1)
def ellipse(x,y,a,b):
    """plots an ellipse given arrays of x and y and values of a and b for the equation (x/a)^2 + (y/b)^2 = 1"""
    plt.figure(figsize=(9,9))
    xp = x[np.where(abs(x)<=a)[0]]
    yp = y[np.where(abs(y)<=b)[0]]
    y_pos = b * np.sqrt( 1 - (xp/a)**2)
    y_neg = -y_pos
    # plotting only y_pos and y_neg leaves gaps near y = 0
    x_pos = a * np.sqrt( 1 - (yp/b)**2) 
    x_neg = -x_pos
    title = 'a = ' + str(a) + '   b = ' + str(b)
    plt.plot(xp, y_pos, color = 'b', linewidth = 3)
    plt.plot(xp, y_neg, color = 'b', linewidth = 3)
    plt.plot(x_pos, yp, color = 'b', linewidth = 3)
    plt.plot(x_neg, yp, color = 'b', linewidth = 3)
    #without manually setting x or y limits the graph will always look like a circle
    if a > b:
        plt.ylim(-a,a)
    if b > a:
        plt.xlim(-b,b)
    plt.xlabel('x', fontsize = 15)
    plt.ylabel('y', fontsize = 15)
    plt.title(title , fontsize = 20 )
ellipse(x,y,12,2)
ellipse(x,y,4,20)
```
