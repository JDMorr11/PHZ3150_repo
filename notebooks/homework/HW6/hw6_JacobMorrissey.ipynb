# Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
```

```python
x = np.arange(1,100.5,.5)
y = np.log10(x)
z = np.sqrt(x)
jtplot.style(theme='onedork')
```

```python
print('Problem 2:')
plt.figure(figsize=(15,15))
plt.plot(x,y,color='red',linestyle='--',linewidth=3, marker = 'o', markersize=2)
plt.plot(x,z,color='forestgreen', linestyle=':',linewidth=2, marker='d', markersize=3)
plt.xlabel('x', fontsize = 25)
plt.ylabel('y and z', fontsize = 25)
plt.xticks(np.arange(0,110,10), fontsize = 20)
plt.yticks(np.arange(0,10.5,.5), fontsize = 20)
plt.tick_params(axis='both',which='major',length=12, width=4,direction='in')
plt.savefig( 'hw6_jamor_prob2_plot.pdf', format = 'pdf', 
            dpi = 1200, bbox_inches = 'tight', pad_inches = 0.3 )
```

```python
word_list = ['twinkle', 'twinkle', 'little', 'star']
for i in word_list:
    if 'a' in i:
        print('a is element', i.index('a'), 'in', i)
    if 'e' in i:
        print('e is element', i.index('e'), 'in', i)
test_words = ['star', 'bright']
for i in test_words:
    if i in word_list:
        print(i, 'is in the list')
    if i not in word_list:
        print(i, 'is not in the list')
```
