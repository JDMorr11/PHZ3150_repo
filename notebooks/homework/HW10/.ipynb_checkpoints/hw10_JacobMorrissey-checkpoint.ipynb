## Problem 2

```python
class holiday_frenzy:
    """Represents a box with dimensions a, b, and c in inches, number of gifts with a volume of 25 cubic inches, and surface area of gift wrap available in square inches."""
    def __init__(self, a = 1, b = 1, c = 1, number_of_gifts = 1, wrap_paper_available = 1): 
        self.a = a
        self.b = b
        self.c = c
        self.number_of_gifts = number_of_gifts
        self.wrap_paper_available = wrap_paper_available
    def surface_area(box):
        s_a = 2 * ( (box.a * box.b) + (box.a * box.c) + (box.b * box.c) )
        return s_a
    def volume(box):
        return (box.a * box.b * box.c)
    def enough_paper(box):
        return box.surface_area() <= box.wrap_paper_available
    def fits(box):
        return box.volume() >= ( 25 * box.number_of_gifts )
    def print_gift(box):
        if box.enough_paper():
            paper = ''
        else:
            paper = 'do not '
        if box.fits():
            gifts = ''
        else:
            gifts = 'not'
        if box.fits() + box.enough_paper() == 1:
            conj = 'but '
        else:
            conj = 'and '
        print('You ' + paper + 'have enough wrapping paper, ' + conj + 'you can' + gifts + ' fit all your gifts in the box.')
        if box.fits() + box.enough_paper() == 0:
            print('The holidays are ruined!')
```

```python
print('Problem 2')
box1 = holiday_frenzy(10,10,10,15,800)
box2 = holiday_frenzy(10,10,10,55,1000)
box3 = holiday_frenzy(100,20,10,55,300)
box1.print_gift()
box2.print_gift()
box3.print_gift()
```

## Problem 3

```python

def piston(V, P0, V0, T0, gamma):
    """Input: an array of volumes, initial values of pressure, volume, and temperature, and gamma. \n Output: an array of P, V, T triples for each input volume, representing an adiabatic expansion."""
    cons1 = P0 * V0 / T0
    cons2 = (P0 * V0) ** gamma
    triple = np.ndarray((len(V), 3))
    for i in range(len(V)):
        P = ( cons2 ** (1 / gamma) ) / V[i]
        T = P * V[i] / cons1
        triple[i,0] = P
        triple[i,1] = V[i]
        triple[i,2] = T
    return triple
```

```python
print('Problem 3')
V = np.linspace(1, 0.1, 40)
V0 = 1
P0 = 1
T0 = 300
gamma = 1.4
triple = piston(V, P0, V0, T0, gamma)
fig = plt.figure( figsize = ( 15, 15 ) )
ax = fig.gca(projection='3d')
ax.plot(triple[:,0], triple[:,1], triple[:,2], linewidth = 4)
ax.set_xlabel('P (atm)')
ax.set_ylabel('V (L)')
ax.set_zlabel('T (K)')
plt.savefig( 'hw10_JaMor_prob3_plot.pdf', format = 'pdf', dpi = 1200, bbox_inches = 'tight', pad_inches = 0.3 )
plt.show()
```
